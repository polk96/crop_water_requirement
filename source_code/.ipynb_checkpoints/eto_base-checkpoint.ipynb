```python
# import eto python library
from eto import ETo, datasets

# import pandas library
import pandas as pd

# Apagamos algunos mensajes de warning
pd.options.mode.chained_assignment = None  # default='warn'
```

```python
# creating object from ETo class
et1 = ETo()
```

```python
# loading path where datasets are stored
#ex1_path = datasets.get_path('example_daily')
ex1_path = '/home/adan/Documents/programacion_analisis_datos/codigos/requerimiento_agua_tomate_rojo/output/conagua_2019_febrero_req_for_et0.csv'
```

```python
# reading example dataset
tsdata = pd.read_csv(ex1_path, parse_dates=True, infer_datetime_format=True, index_col='date')
```

```python
tsdata.info()
```

```python
# parameters definition
# high meter sea level in meters or altitude
z_msl = 1408.99

# The latitude of the met station (dec deg) (only needed if R_s or R_n are not in df).
lat = 19.59216667

# The longitude of the met station (dec deg) (only needed if calculating ETo hourly)
lon = -103.59075

# The longitude of the center of the time zone 
# (dec deg) (only needed if calculating ETo hourly).
TZ_lon = -103

# The Pandas time frequency string of the input and output. 
# The minimum frequency is hours (H) and the maximum is month (M)
freq = 'D'
```

```python
# loading base parameters so rest of them can be estimated automatically
et1.param_est(tsdata, freq, z_msl, lat, lon, TZ_lon)
```
